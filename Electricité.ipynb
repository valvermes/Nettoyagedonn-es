# On retire les autres caractères de la première colonne (on veut dans SITE_Name que les 4 chiffres et les deux caractères

```python
df['SITE'] = df['SITE_NAME'].str[11:17]
```

# Ajoute une colonne qui calcule le temps entre Begin et END

```python
import datetime

df[['BEGIN','END']] = df[['BEGIN','END']].apply(pd.to_datetime)

df
```

```python
df['DUREE']=(df['END']-df['BEGIN']).dt.days.astype('Int64')
```

```python
df
```

# Colonne consommation journalière moyenne de la période

```python
df.dtypes
```

# Il y a un espace dans le string donc je l'enlève

```python
df['KWH'] = df['KWH_PERIOD'].str[0:1] + df['KWH_PERIOD'].str[2:]
```

```python
df
```

```python
df.dtypes
```

```python
df['KWH'] = pd.to_numeric(df['KWH'], errors='coerce').fillna(0).astype(np.int64)
```

```python
df.dtypes
```

```python
df
```

```python
df['CONSOMMATION'] = df['KWH'] / df['DUREE']
```

```python
df
```
